### Read and Save .csv Files

Various commonly used file formats can be read using Python, such as .csv and .xls files. You can read these files by using the pandas library. Firstly, the following code must be used to access the pandas library:

````python
import pandas as pd
````

Having run this (Shift+enter or Ctrl+enter), it is now possible to read in a .csv file data using the following command:

```
DataFrame Object = pd.read_csv("filepath and name")
```

The filepath and name are enclosed in quotes with the .csv (or appropriate) extension. Please note that for file path you need to use forward slash / instead of back slash \. 

**Example:**

````python
borders = pd.read_csv("H:/Python Training Data/Borders.csv")
````

*Note: using an = assigns a value (i.e. the text on the right of the = sign) to a dataframe object (i.e. the text on the left of the = sign). In the above example, the dataframe Borders.csv has been assigned to the dataframe object borders. Remember that Python is a case-sensitive language.*

*Variables (columns in the dataframe) and their data types do not need to be explicitly declared, as Python does this automatically when importing the dataframe.*

To view the dataframe in the console, you will have to put an extra line of code stating the dataframe object i.e. borders. However, if it is a large dataframe and you only need to check the data columns quickly by looking at the first few rows, you can stipulate the number of rows you would like to display using the following command:

```
df.head(n)
```

n is the number of rows you want to display. If you dont specify n, it will default as 5 rows. 

**Example:**

````python
borders.head(10)
````

In most cases, once a dataframe has been imported, you are likely to make changes e.g. adding new columns or changing values in existing columns. If you wish to keep the changed data for later use, you have two choices:

1. Save the python code that you have created, and then re-run this code on the original file at a later date.
2. Save the dataframe in its updated format as a new file.

Option 1 is a good choice if the original file will not change in the intervening period and the changes made to the file using the code do not take a long time to execute. If neither of these conditions apply, the second option is often quicker and easier.

The simplest way to save a new .csv file is to use the ```df.to_csv``` command ( df is the dataframe object e.g. borders in the current example):

```
df.to_csv(filepath and name)
```

**Example:**

````python
borders.to_csv("H:/Python Training Data/Borders2.csv")
````

### Data Structure and Missing Data

Once dataframe has been imported (i.e. read in), it is important to undertake simple exploration. By carrying out basic checks and getting to know the data, any problems or errors in the data can be discovered and dealt with before they become a problem in later analysis.

As a first step in this process, it is important to know the size of the dataframe. This can be determined by using the following command:

```
df.shape
```

**Example:**

We can check how many rows and columns there are in the dataframe object borders, which was created in Section 2.1. The output shows it has 25299 rows and 14 columns.

````python
borders.shape
````

You may also have  missing values in your dataframe. These will be displayed as NaN (i.e. not a number). The following command will print the dataframe in the console and indicate which cells have missing values:

```
df.isnull()
```

**Example:**

We can check the missing values for dataframe object borders. Each cell will show True (is missing value) or False (is not a missing value). 

````python
borders.isnull().head()
````

However, for large dataframes, it may be quicker to see which columns have missing values.  You can determine this by using the following command:

```
df.isnull().any()
```
**Example:**

If we want to check if any value in a column contains missing value for borders dataframe object, we can use the following command and the outputs will show True (contains missing values) or False (does not contain missing values). 

````python
borders.isnull().any()
````

Note: As we have seen in the above examples, the ```df.isnull()``` and ```df.null.any()``` commands return output in the form of Boolean (or logical) expressions. To this end, Python provides the Boolean type that can either be set to **true** or **false**. Many command and functions that you will use in Python return Boolean results. 

To change missing (i.e. NaN) values to 0, you can use the following command:

```    
df.fillna(0)
```
To remove rows with missing values, you can use the following command:

```
df.dropna()
```

More methods of dealing with [missing values](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html) can be found in the Pandas documentation.

### Frequencies and Crosstabs

Performing frequencies and cross tabulations on relevant variables once data have been imported are good first steps in data exploration.
For example, you might check that:

* You have the right dataset e.g. by checking it has the right number of records and covers the right time period. 

* Columns contain no unusual values and the values appear in the expected proportions (e.g. you may be expecting roughly 50/50 males and females). 

* Cross-tabulating reveals no unusual combinations of values e.g. a 'Male' person who has had a 'Hysterectomy'! Checks such as these usually require an understanding of background information regarding the data. 

* There are no missing values. 

Performing such checks often will help you to avoid problems later on in the analysis. Sometimes problems in your data that were identifiable at this stage are not identified until data has been sent to the customer!

````python
borders['Specialty'].value_counts()
````

In addition to basic frequencies, it is also possible to calculate the mean and median values:


<a id = 'select-rows'></a>
### Select Rows

In addition to selecting specific columns, you may also be interested in selecting a subset of rows in your dataframe. This can be done in several ways. Some of the most common methods are outlined in the following sections.

#### Indexing

Select rows using row index. It is important to know that the row index starts from 0 in Python. You can use the code

```
df[row index x : row index y]
```

Please note that the selection of rows will include row index x but not include row index y.

**Example:**

In this example we would like to select the first 5 rows. To do this, it is necessary to input 0 (the first row that will be included) as row index x and 5 (the 6th row that will not be included) as row index y in the command.

````python
borders[0:5]
````

If row index x is omitted, it will be defaulted as 0. For example, ```df[:10]``` will take the row index from 0 to 9 (not including 10). 

If you want to get the bottom of 10 rows, you can use negative number for row index x as ```df[-10:]```.

#### Using String Key Words

You can have columns in a dataframe showing as characters, such as hospital names. Such columns are called string columns. For a string column, we can select cases based on certain key words in that string. We first need to create a boolean object with values either True or False to flag if each case contains that key word, and then select these cases from the dataframe.

**Example:**

We want to select the cases with specialty containing E. The code is:

````python
bor_specialty = borders['Specialty'].str.contains('E')
borders[bor_specialty].head()
````

The first command line created a new object called bor_specialty with values True or False judging by whether Specialty contains character E. The second command used bor_specialty as a condition in the square bracket to filter out the desired cases. 

Thus you can see all the cases with specialty containing E are selected.

<a id = 'using-conditions'></a>
#### Using Conditions

It is possible to select rows in a dataframe using conditional statements. The table below shows a selection of comparison operators that can be used in conditional statements.

| Operator | Definition |
| :--- | --- |
| == | Equal to |
| != | Not equal to |
| < | Less than |
| <= | Less than or equal to |
| > | Greater than |
| >= | Greater than or equal to |

To select rows in Python using conditional statements, the following command is needed:

```
df[df['column name'] conditional statement]
```

**Example:**

In this example, we are only interested in cases relating to the Borders hospital. Borders hospital is denoted by the code B120 in the HospitalCode column. Therefore, to select only rows relating to Borders hospital, we use the command below (note the appended ```head()``` method will show the first 5 rows).

````python
borders[borders['HospitalCode'] == "B120H"].head()
````

In addition to comparison operators, sometimes it is necessary to use logical operators when selecting rows using conditional statements. The table below presents some of the most common logical operators.

| Operator | Definition |
| :--- | --- |
| And (&) | Both relations must be true |
| Or (&#124;) | Either relation can be true |
| Not | Reverses the outcome of an expression |

**Example:**

In this example, we would like to select cases for Borders General (B120H) and cases for Kelso Hospital (B114H).

````python
borders[(borders['HospitalCode'] == "B120H")|(borders['HospitalCode'] == "B114H")].head()
````

There is an alternative method to complete the above task. This involves the creation of Boolean objects which are TRUE under specific conditions. Here we use hash key # to add comments. It is a good habit to write comments to keep a note of what things the following codes will do. 

**Example:**

````python
# Create an object with TRUE if hospital code is equal to B120H.
BG = borders['HospitalCode'] == "B120H" 
# Create an object with TRUE if hospital code is equal to B114H.
K = borders['HospitalCode'] == "B114H"
# Select rows where hospital code is B120H or B114H
borders[BG|K].head()
````

### Merge Dataframe

Pandas library also makes it easy to merge dataframes by matching two files together by certain key variables. For example, sometimes a reference or lookup table is used to add variables to all cases with matching values for the key variables in the lookup.
```
pd.merge(dataframe1, dataframe2, how = "left", on = ['key variable1', 'key variable2'])
```

dataframe2 refers to the lookup dataset, whilst dataframe1 refers to the main file that you want to add information on. Lets have a look at an example. Here we will use Baby5.csv and Baby6.csv. We want to find babies surname from Baby6.csv and add the information to Baby5.csv.

### Format Data

When reading data you might want to reformat certain columns such as dates or decimals. This can be achieved through some python functions which are illustrated below. Generally data only needs to be formatted if you are merging files (to ensure data contained in both files is all of the same type and format)  or outputting data (such as exporting your data to a final Excel output).  

#### Formatting Dates

Formatting dates is a common method of data manipulation in Python. The default format for dates in Python is YYYY-MM-DD. 

To format dates we first need to import datetime library, and then we can use functions ```strptime()``` and ```strftime()```. These functions allow you to change the appearance of dates. The ```strptime()``` function takes a given character date (i.e. 23/06/2017) and outputs this to a standard date format. For example, the following code will take this character value, specify its format and then output this as a standard date:

````python
import datetime
date1 = datetime.datetime.strptime("23/06/2017", "%d/%m/%Y")
print(date1)
````

Running this code takes the character value and outputs it in a standard YYYY-MM-DD format. The code ```format = "%d/%m/%Y"``` specifies the format of the character value you wish to transform into a date value. The %d denotes the day of the month as a number (01-31), %m denotes the month as a number (01-12) and %Y denotes the full year including the century.

The ```strftime()``` function works in the opposite way to ```strptime()```. Whereas ```strptime()``` takes a character value and converts it to a date value, ```strftime()``` performs the reverse action. A simple way to test this function is by using the ```datetime.date.today()``` function. This function outputs todays date in standard date format. Then we can use ```strftime()``` to change its format.

````python
today = datetime.date.today().strftime("%d-%m-%Y")
print(today)
````

Note that the date format in ```strftime()``` still refers to the character value, but this time it specifies the desired output for the character value. 

However, when you deal with the date column in a dataframe, the easiest way to transform it into standard date format (YYYY-MM-DD) is to use pandas function ```pd.to_datetime()```. If you want to use ```strptime()```, you have to use function ```apply()``` outside ```strptime()```, as the date we want to transform is not a single string but a series of dates in that whole column. 

**Example:**

We want to transform admissiondate in BORDERS (inc Age).csv which is currently in format MM/DD/YYYY into standard date format YYYY-MM-DD. 

````python
import pandas as pd
borders_age = pd.read_csv("H:/Python Training Data/BORDERS (inc Age).csv")
borders_age.head()
````

````python
borders_age['admissiondate'] = pd.to_datetime(borders_age['admissiondate'], format = "%m/%d/%Y")
borders_age['admissiondate'].head()
````

Alternatively, You can also use ```strptime()``` with ```apply()```:

````python
borders_age['admissiondate'] = borders_age['admissiondate'].apply(lambda x: datetime.datetime.strptime(x, "%m/%d/%Y"))
borders_age['admissiondate'].head()
````

After doing that, you can transform it to another format such as DD-MM-YYYY using ```strftime()``` with ```apply()```:

````python
borders_age['admissiondate'] = borders_age['admissiondate'].apply(lambda x: datetime.datetime.strftime(x, "%d/%m/%Y"))
borders_age['admissiondate'].head()
````

**Example:**

We want to change column A in NumberFormat.csv from numeric to string:

#### Formatting Decimals

Sometimes you want to format the numbers by a certain number of decimal places. This can be achieved by using function ```round()```.

**Example:**

We want to format all the numbers in NumberFormat.csv to have 2 decimal places:

````python
import pandas as pd

NumberFormat = pd.read_csv("H:/Python Training Data/NumberFormat.csv")

NumberFormat.head()
````

````python
NumberFormat.round(2)
````

If we want to format column A to have 1 decimal place whilst column C to have 2 decimal places, we can run the following code:

````python
NumberFormat.round({'A':1, 'C':2})
````

It should be noted that rounding in Python is slightly different to SPSS. Whereas SPSS rounds down for values < 5 and rounds up for values >= 5, the ```round()``` function uses a different rounding technique. Python uses a round to the nearest even integer method, i.e. 2.5 would round down to 2 rather than 3, whereas 3.5 would round up to 4. 

As a result, care should be taken when comparing results for rounded numbers between Python and SPSS. Please read the guidance paper [Rounding Methods in Different Software](https://www.isdscotland.org/About-ISD/Methodologies/_docs/Rounding-Methods-in-Different-Software_v1-0.pdf) for more information.

#### Adding in Leading Zeros

You might have to add leading zeros to some of your data to ensure it is in the required format (e.g. CHI number). The most straight forward way of doing this is ```str.zfill()```. You need to put a number of string length you wish it should become in the parantheses. 

**Example:**

Using the DOCTORS.csv file, add leading zeros to the "dobmm" column such that each row consists of two numbers. Remeber to change the data type from integer to string first. 

````python
import pandas as pd

DOCTORS = pd.read_csv("H:/Python Training Data/DOCTORS.csv")

# Transform the data type from integer to string
DOCTORS['dobmm']  = DOCTORS['dobmm'].astype(str)

# Fill the string with a leading zero
DOCTORS['dobmm']  = DOCTORS['dobmm'].str.zfill(2) 
DOCTORS.head()
````

The "dobmm" column has now been filled out to add leading zeros where required.

### Plot with Matplotlib

The simplest approach to producing plots with matplotlib is to use the pyplot package. This package is contained within the matplotlib library and is very useful for creating straight forward graphics. Pyplot uses a series of simple Python functions to generate plots in a quick and efficient manner.

The pyplot package can be imported using the code detailed below. The name plt is a commonly used term for the pyplot package. 

````python
import matplotlib.pyplot as plt
````

An easy way to print the plot to the console is to generate the plot inline within Jupyter notebook. To set this up, we need to run the following code:

```
%matplotlib inline
```

An example of an inline plot is shown below. This code creates a simple line graph.

````python
%matplotlib inline

# Plots a line plot
plt.plot([1, 3, 2, 4])
````

We can also create a bar plot and have this appear directly beneath the relevant code within Jupyter notebook. 

In the example below, the Borders (inc Age).csv dataset is read in and a bar plot is created for the frequency of each admission day. The ```value_counts()``` function calculates the frequency for each admission day, while ```sort_index()``` arranges the frequencies in order of their admission day. Finally, ```plot.bar()``` produces the bar plot shown below the code. 

````python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# Read in data
borders_age = pd.read_csv('H:/Python Training Data/Borders (inc Age).csv')

# Plot a bar chart for the frequency of each admission day
borders_age['admissionday'].value_counts().sort_index().plot.bar()

# Show the plot
plt.show()
````

The bar plot is shown below the code used to create it. This can be very useful, especially if the plot is being created for some quick exploratory analysis, as it gives a clear and immediate insight into certain characteristics of the dataframe. 

<a id = 'add-titles-and-axis-labels'></a>
### Add Titles and Axis Labels

While the bar plot we created above has labels for each bar, it doesnt have a title or axis labels. Thus, to anyone without access to the data set this plot would be unusable. The pyplot package allows for titles and labels to easily be inserted into charts. 

````python
%matplotlib inline

borders_age['admissionday'].value_counts().sort_index().plot.bar()

# Add a title
plt.title('Frequency of Admission Day')

# Add axis labels
plt.xlabel('Admission Day')
plt.ylabel('Frequency')

plt.show()
````

The plot has now been updated to include a title and labels on both axes.

<a id = 'customise-plots'></a>
### Customise Plots

It is possible to customise plots created with Python. A couple of easy ways to make your plot look more appealing is to change the colour and size. This can be done within the plot function. 

We can easily change the colour of the bar plot as well as adjust the width of the bars. The color command allows you to specify a colour, while width sets the width of the bars. 

Another change we can make in Python is editing the appearance of tick labels. These are the labels shown on each axis that represent each level of the chart, i.e. 1-7 on the x axis in the admission day chart. However, you will notice that the numbers are shown on their side. We can easily make them appear upright again using the ```plt.xticks()``` function and changing their rotation.

````python
%matplotlib inline

# Adjust the colour and width of the bars within the plot.bar() function
borders_age['admissionday'].value_counts().sort_index().plot.bar(color = 'green', width = 0.8)

# You can specify a rotation for the tick labels
plt.xticks(rotation = 0)

# Add a title
plt.title('Frequency of Admission Day')

# Add axis labels
plt.xlabel('Admission Day')
plt.ylabel('Frequency')

plt.show()
````

<a id = 'save-the-output'></a>
### Save the Output

The bar plot above can be saved by using the ```plt.savefig()``` function. This function requires a file path and name to save to and also contains some other options to customise the plot upon saving. One useful option is to use the bbox_inches = 'tight' command. This will ensure that the area of the plot is kept to a minimum when saved, such that all objects in the figure are included. 

````python
plt.savefig('H:/myplot.png', bbox_inches = 'tight')
````

It is important to note that the ```plt.savefig()``` function must be ran before ```plt.show()```. If you use ```plt.show()``` before ```plt.savefig()```, Python will save a blank image. For example, you would need to use the following code if you wanted to use both functions:

````python
plt.savefig('H:/myplot.png', bbox_inches = 'tight')

plt.show()
````
